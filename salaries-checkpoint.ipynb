```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('salaries.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_company = LabelEncoder()
label_job = LabelEncoder()
label_degree = LabelEncoder()
```

```python
df['label_company'] = label_company.fit_transform(df['company'])
df['label_job'] = label_company.fit_transform(df['job'])
df['label_degree'] = label_company.fit_transform(df['degree'])
```

```python
df.head()
```

```python
input_value = df.drop(['company','job','degree','salary_more_then_100k'], axis = 1)
target = df['salary_more_then_100k']
```

```python
input_value.head()
```

```python
target.head()
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(input_value,target)
```

```python
model.predict([[2,2,1]])
```

```python
plt.figure(figsize=(6,6))
tree.plot_tree(model, filled=True)
```
